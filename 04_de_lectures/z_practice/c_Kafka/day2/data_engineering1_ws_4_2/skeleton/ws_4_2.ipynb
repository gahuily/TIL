```python
"""
Kafka 프로듀서를 활용하여 메시지를 Kafka 토픽으로 전송하는 실습입니다.

TODO:
1. kafka-python 라이브러리를 사용하여 KafkaProducer를 생성합니다.
2. test-topic으로 메시지를 전송합니다.
3. 여러 개의 메시지를 반복적으로 전송하는 기능을 구현합니다.
"""

from kafka import KafkaProducer
```

```python
# TODO 1: KafkaProducer를 생성하세요.
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',  # TODO: Kafka 브로커 주소를 설정하세요.
    key_serializer=str.encode, # 키 직렬화 방식 설정
    value_serializer=str.encode # 값 직렬화 방식 설정
)
```

```python
# TODO 2: test-topic으로 메시지를 전송하세요.
topic = "test-topic"
message = 'Hello Kafka'  # TODO: 전송할 메시지를 설정하세요.
```

```python
# TODO 3: 여러 개의 메시지를 반복적으로 전송하는 기능을 구현하세요.
for i in range(5):
    producer.send(topic, key=f"key-{i}", value=f"{message}-{i}")  # 프로듀서에서 메시지를 보내는 함수
    print(f"Sent: key-{i}, message-{i}")
```
