```python
"""
Kafka 프로듀서를 활용하여 test-topic을 삭제 후 재생성하고, 메시지를 전송하는 실습입니다.

TODO:
1. kafka-python 라이브러리를 사용하여 KafkaAdminClient를 생성합니다.
2. 기존 test-topic이 존재하면 삭제한 후, 2초 대기 후 새 토픽을 생성합니다.
3. KafkaProducer를 사용하여 test-topic으로 메시지를 전송합니다.
4. 여러 개의 메시지를 반복적으로 전송하는 기능을 구현합니다.
"""

import time
from kafka import KafkaProducer
from kafka.admin import KafkaAdminClient, NewTopic
```

```python
# TODO 1: KafkaAdminClient를 생성하세요.
admin_client = KafkaAdminClient(
    bootstrap_servers='localhost:9092',  # TODO: Kafka 브로커 주소를 설정하세요.
    client_id="producer-admin"
)

topic_name = "test-topic"
```

```python
# TODO 2: 기존 test-topic을 삭제하고, 2초 대기 후 파티션 개수가 2개인 새 토픽을 생성하세요.
existing_topics = admin_client.list_topics()  # 기존 토픽 목록을 가져오는 함수
if topic_name in existing_topics:
    admin_client.delete_topics([topic_name])  # 특정 토픽을 삭제하는 함수
    print(f"기존 토픽 '{topic_name}' 삭제 완료")
    time.sleep(2)  # 토픽 삭제 후 2초 대기 (삭제 완료 대기)

day2_new_topic = NewTopic(name=topic_name, num_partitions=2, replication_factor=1)
admin_client.create_topics([day2_new_topic])  # 새로운 토픽을 생성하는 함수
print(f"새로운 토픽 '{topic_name}' 생성 완료")

admin_client.close()  # AdminClient 연결을 닫는 함수
```

```python
# TODO 3: KafkaProducer를 생성하세요.
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',  # TODO: Kafka 브로커 주소를 설정하세요.
    value_serializer=str.encode
)
```

```python
# TODO 4: test-topic으로 메시지를 전송하세요.
for i in range(100):
    producer.send(topic_name, value=f"message-{i}")  # 프로듀서에서 메시지를 보내는 함수
    print(f"Sent: message-{i}")
```
